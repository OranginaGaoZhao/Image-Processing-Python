```python
%matplotlib inline
import matplotlib.pyplot as plt
from scipy import misc
test_image = misc.imread('wallpaper.jpg')
```

```python
test_image.shape, test_image.dtype
```

```python
plt.imshow(test_image)
plt.axis('off')
plt.title('Wallpaper')
```

```python
downsample_image = misc.imresize(test_image, 1./3, 'bicubic')
```

```python
plt.imshow(downsample_image)
plt.axis('off')
plt.title('Downsampled Wallpaper')
misc.imsave('test.png', downsample_image)
```

```python
crop_image = test_image[100:800,100:800,:]
plt.imshow(crop_image)
plt.axis('off')
plt.title()
misc.imsave('test.bmp', crop_image)
```
